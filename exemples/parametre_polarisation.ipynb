```python
import fonction_polarisation as fp
import numpy as np
import matplotlib.pyplot as plt
```

```python
angle = 160*np.pi/180
a = np.sqrt(0.8)
b = np.sqrt(0.2)
ellipse = np.arctan(b/a)
print("angle:", angle*180/np.pi, "ellipse:", ellipse)

vect_jones_0 = fp.j_theta_epsilon(angle, ellipse)
print(vect_jones_0)
fp.graph_j(vect_jones_0)
```

```python
psi, delta = fp.psidelta_j(vect_jones_0)
print("psi:", psi*180/np.pi, "delta:", delta*180/np.pi)
```

```python
# Generate some example data
theta = np.linspace(0, 2*np.pi, 100)  # Angle values
print(np.real(np.dot(np.conjugate(vect_jones_0), vect_jones_0)))

r = [0 for i in range(100)]# Radial values

for i in range(100):
    r[i] = np.real(np.dot(np.conjugate(np.dot(fp.polariseur(theta[i]), vect_jones_0)), np.dot(fp.polariseur(theta[i]), vect_jones_0)))

# Create a polar plot
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)  # Create a polar subplot
# Plot the data on the polar plot
ax.plot(theta, r)
# Customize the polar plot
ax.set_rlabel_position(90)  # Set radial labels position
#ax.set_rticks([])  # Remove radial tick labels
ax.spines['polar'].set_visible(False)  # Remove the polar plot spine
plt.show()
```
